```python
import geopandas as gpd
```

```python
pop = gpd.read_file("assets/2020population_4326.geojson")
stops = gpd.read_file("assets/data/stops/link_stops.geojson")
```

```python
pop = pop.to_crs(stops.crs)
```

```python
joined = gpd.sjoin(pop, stops, how="inner", predicate="contains")
```

```python
dissolved = joined.dissolve(by="id")
```

```python
dissolved.to_file("population_by_link_stop.geojson", driver="GeoJSON")
```
